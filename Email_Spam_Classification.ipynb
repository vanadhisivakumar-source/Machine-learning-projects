# Task
Implement a Gaussian Naïve Bayes classifier to determine if a new email with features "Contains Free = Yes", "Contains Win = Yes", and "Contains Offer = No" is spam or not spam.

## Load and Prepare Data

### Subtask:
Create a pandas DataFrame from the provided dataset and encode 'Yes'/'No' values to numerical representations.


**Reasoning**:
First, I'll define the dataset as a dictionary, then convert it into a pandas DataFrame as instructed.



```python
import pandas as pd

data = {
    'Contains Free': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes'],
    'Contains Win': ['No', 'Yes', 'No', 'No', 'Yes', 'No', 'No', 'No', 'Yes', 'Yes'],
    'Contains Offer': ['No', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'No', 'No', 'Yes'],
    'Spam': ['No', 'Yes', 'No', 'No', 'Yes', 'No', 'No', 'No', 'Yes', 'Yes']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df.head())
```

**Reasoning**:
Now that the DataFrame is created, I will encode the 'Yes'/'No' values to numerical representations (1 for 'Yes' and 0 for 'No') in all relevant columns as per the instructions.



```python
df_encoded = df.replace({'Yes': 1, 'No': 0})
print("\nEncoded DataFrame:")
print(df_encoded.head())
```

**Reasoning**:
The previous code block failed because the DataFrame 'df' was not defined in the current execution context. To resolve this, I will combine the DataFrame creation and the encoding steps into a single code block to ensure 'df' is available when 'df.replace()' is called.



```python
import pandas as pd

data = {
    'Contains Free': ['Yes', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes'],
    'Contains Win': ['No', 'Yes', 'No', 'No', 'Yes', 'No', 'No', 'No', 'Yes', 'Yes'],
    'Contains Offer': ['No', 'No', 'Yes', 'No', 'No', 'Yes', 'No', 'No', 'No', 'Yes'],
    'Spam': ['No', 'Yes', 'No', 'No', 'Yes', 'No', 'No', 'No', 'Yes', 'Yes']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df.head())

df_encoded = df.replace({'Yes': 1, 'No': 0})
print("\nEncoded DataFrame:")
print(df_encoded.head())
```

**Reasoning**:
The previous step successfully created and encoded the DataFrame. To prepare the data for training a Gaussian Naïve Bayes classifier, the next logical step is to separate the features (input variables) from the target variable ('Spam') into X and y, respectively.



```python
X = df_encoded.drop('Spam', axis=1)
y = df_encoded['Spam']

print("Features (X):")
print(X.head())
print("\nTarget (y):")
print(y.head())
```

## Calculate Prior Probabilities

### Subtask:
Calculate the prior probabilities for the 'Spam' and 'Not Spam' classes (P(Spam=1) and P(Spam=0)) from the prepared training data.


**Reasoning**:
To calculate the prior probabilities, I will count the occurrences of 'Spam' (1) and 'Not Spam' (0) in the target variable `y` and divide by the total number of instances. This will provide P(Spam=1) and P(Spam=0).



```python
total_instances = len(y)

# Calculate count of spam and not spam emails
spam_count = y.sum()  # y contains 1s for spam
not_spam_count = total_instances - spam_count # y contains 0s for not spam

# Calculate prior probabilities
p_spam = spam_count / total_instances
p_not_spam = not_spam_count / total_instances

print(f"Total instances: {total_instances}")
print(f"Count of Spam emails: {spam_count}")
print(f"Count of Not Spam emails: {not_spam_count}")
print(f"Prior probability of Spam (P(Spam=1)): {p_spam:.2f}")
print(f"Prior probability of Not Spam (P(Spam=0)): {p_not_spam:.2f}")
```

## Calculate Gaussian Parameters

### Subtask:
For each feature ('Contains Free', 'Contains Win', 'Contains Offer') and for each class ('Spam=1' and 'Spam=0'), calculate the mean and standard deviation of the feature values.


**Reasoning**:
To calculate the Gaussian parameters (mean and standard deviation) for each feature per class, I will first filter the feature set `X` based on the corresponding `y` values (Spam=1 and Spam=0). Then, I will compute the mean and standard deviation for each feature within these filtered datasets. Finally, I'll print these statistics for review.



```python
X_spam = X[y == 1]
X_not_spam = X[y == 0]

mean_spam = X_spam.mean()
std_spam = X_spam.std()

mean_not_spam = X_not_spam.mean()
std_not_spam = X_not_spam.std()

print("\n--- Statistics for SPAM (y=1) ---")
print("Mean (Spam):\n", mean_spam)
print("Standard Deviation (Spam):\n", std_spam)

print("\n--- Statistics for NOT SPAM (y=0) ---")
print("Mean (Not Spam):\n", mean_not_spam)
print("Standard Deviation (Not Spam):\n", std_not_spam)
```

## Define Gaussian PDF Function

### Subtask:
Implement a Python function that calculates the probability density using the Gaussian (Normal) distribution formula, given a value, mean, and standard deviation.


**Reasoning**:
To implement the Gaussian PDF function as instructed, I will define a Python function that calculates the probability density using the Gaussian formula. This function will import the `math` module, take `x`, `mu`, and `sigma` as inputs, and include a check for a zero standard deviation to prevent division by zero errors.



```python
import math

def gaussian_pdf(x, mu, sigma):
    """
    Calculates the probability density using the Gaussian (Normal) distribution formula.

    Args:
        x (float): The value at which to evaluate the PDF.
        mu (float): The mean of the distribution.
        sigma (float): The standard deviation of the distribution.

    Returns:
        float: The probability density at x.
    """
    # Add a small epsilon to sigma to prevent division by zero or issues with logarithm
    # This handles cases where all feature values for a given class are identical.
    epsilon = 1e-9
    sigma = max(sigma, epsilon)

    coefficient = 1 / (math.sqrt(2 * math.pi * sigma**2))
    exponent = -((x - mu)**2 / (2 * sigma**2))
    pdf = coefficient * math.exp(exponent)
    return pdf

print("gaussian_pdf function defined.")
```

## Define New Email Features

### Subtask:
Define the features of the new email ('Contains Free', 'Contains Win', 'Contains Offer') that needs to be classified as 'Spam' or 'Not Spam'.

**Reasoning**:
The subtask requires defining the new email's features. Following the encoding scheme established earlier ('Yes' to 1, 'No' to 0), I will create a dictionary representing these features, convert it to a pandas Series for consistency with the model's expected input format, and then print it for verification.



```python
new_email_features_raw = {
    'Contains Free': 'Yes',
    'Contains Win': 'Yes',
    'Contains Offer': 'No'
}

# Encode the new email features using the same mapping as the training data
new_email_features = pd.Series({
    'Contains Free': 1 if new_email_features_raw['Contains Free'] == 'Yes' else 0,
    'Contains Win': 1 if new_email_features_raw['Contains Win'] == 'Yes' else 0,
    'Contains Offer': 1 if new_email_features_raw['Contains Offer'] == 'Yes' else 0
})

print("New email features (encoded):")
print(new_email_features)
```

## Calculate Likelihood for Each Class

### Subtask:
Calculate the likelihoods P(Features | Spam) and P(Features | Not Spam) for the new email using the `gaussian_pdf` function and the previously calculated means and standard deviations.

**Reasoning**:
To calculate the likelihood for each class, I will iterate through each feature of the new email, use the previously defined `gaussian_pdf` function with the respective means and standard deviations for 'Spam' and 'Not Spam', and multiply these probabilities together for each class to get the overall likelihood.



```python
likelihood_spam = 1.0
likelihood_not_spam = 1.0

for feature in new_email_features.index:
    x_val = new_email_features[feature]

    # Likelihood for Spam class
    mu_spam = mean_spam[feature]
    sigma_spam = std_spam[feature]
    pdf_spam = gaussian_pdf(x_val, mu_spam, sigma_spam)
    likelihood_spam *= pdf_spam

    # Likelihood for Not Spam class
    mu_not_spam = mean_not_spam[feature]
    sigma_not_spam = std_not_spam[feature]
    pdf_not_spam = gaussian_pdf(x_val, mu_not_spam, sigma_not_spam)
    likelihood_not_spam *= pdf_not_spam

print(f"\nLikelihood P(Features | Spam): {likelihood_spam}")
print(f"Likelihood P(Features | Not Spam): {likelihood_not_spam}")
```

## Predict for New Email

### Subtask:
Using the calculated prior probabilities and the Gaussian parameters, apply the Naïve Bayes formula to compute P(Spam|X) and P(NotSpam|X) for the new email (Contains Free = Yes, Win = Yes, Offer = No). Compare these probabilities to determine the predicted class.


**Reasoning**:
To complete the subtask, I will calculate the unnormalized posterior probabilities for both 'Spam' and 'Not Spam' classes using the likelihoods and prior probabilities. Then, I'll compare these values to determine the predicted class for the new email.



```python
posterior_spam_unnorm = likelihood_spam * p_spam
posterior_not_spam_unnorm = likelihood_not_spam * p_not_spam

print(f"\nUnnormalized Posterior P(Spam | Features): {posterior_spam_unnorm}")
print(f"Unnormalized Posterior P(Not Spam | Features): {posterior_not_spam_unnorm}")

# Compare the unnormalized posterior probabilities
if posterior_spam_unnorm > posterior_not_spam_unnorm:
    predicted_class = 'Spam'
else:
    predicted_class = 'Not Spam'

print(f"\nPredicted class for the new email: {predicted_class}")
```
