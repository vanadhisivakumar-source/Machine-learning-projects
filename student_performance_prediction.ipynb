```python
import pandas as pd

# Create the dataset
data = {
    'StudyHours': ['High', 'High', 'Low', 'Low', 'High', 'Low'],
    'Attendance': ['High', 'Low', 'High', 'Low', 'High', 'Low'],
    'Pass': ['Yes', 'Yes', 'Yes', 'No', 'Yes', 'No']
}
df = pd.DataFrame(data)

display(df)
```

### Probability Tables Calculation

```python
# Calculate Prior P(Pass)
p_pass_yes = df['Pass'].value_counts(normalize=True).get('Yes', 0)
p_pass_no = df['Pass'].value_counts(normalize=True).get('No', 0)

print(f"P(Pass=Yes) = {p_pass_yes:.2f}")
print(f"P(Pass=No) = {p_pass_no:.2f}")

# Calculate Conditional P(StudyHours | Pass)
p_studyhours_given_pass = df.groupby('Pass')['StudyHours'].value_counts(normalize=True)

print("\nP(StudyHours | Pass):")
print(p_studyhours_given_pass)

# Calculate Conditional P(Attendance | Pass)
p_attendance_given_pass = df.groupby('Pass')['Attendance'].value_counts(normalize=True)

print("\nP(Attendance | Pass):")
print(p_attendance_given_pass)
```

### Example Inference: Given StudyHours=High, Attendance=High

```python
# Based on the calculated probabilities from the dataset:
# P(Pass=Yes) = 0.67
# P(Pass=No) = 0.33

# P(StudyHours=High | Pass=Yes) = 0.75
# P(StudyHours=High | Pass=No) = 0.00 (since no 'High' StudyHours when 'Pass' is 'No')

# P(Attendance=High | Pass=Yes) = 0.75
# P(Attendance=High | Pass=No) = 0.00 (since no 'High' Attendance when 'Pass' is 'No')

# P(Pass | Data) \propto P(Pass) * P(StudyHours | Pass) * P(Attendance | Pass)

# For Pass = Yes
posterior_pass_yes = p_pass_yes * \
                     p_studyhours_given_pass.get(('Yes', 'High'), 0) * \
                     p_attendance_given_pass.get(('Yes', 'High'), 0)

# For Pass = No
posterior_pass_no = p_pass_no * \
                    p_studyhours_given_pass.get(('No', 'High'), 0) * \
                    p_attendance_given_pass.get(('No', 'High'), 0)

print(f"P(Pass=Yes | StudyHours=High, Attendance=High) \propto {posterior_pass_yes:.4f}")
print(f"P(Pass=No | StudyHours=High, Attendance=High) \propto {posterior_pass_no:.4f}")

# To get normalized probabilities, sum the unnormalized posteriors
total_posterior = posterior_pass_yes + posterior_pass_no

normalized_p_pass_yes = posterior_pass_yes / total_posterior if total_posterior else 0
normalized_p_pass_no = posterior_pass_no / total_posterior if total_posterior else 0

print(f"\nNormalized P(Pass=Yes | StudyHours=High, Attendance=High) = {normalized_p_pass_yes:.4f}")
print(f"Normalized P(Pass=No | StudyHours=High, Attendance=High) = {normalized_p_pass_no:.4f}")

if normalized_p_pass_yes > normalized_p_pass_no:
    prediction = "Yes"
else:
    prediction = "No"

print(f"\nPrediction \u2192 Pass = {prediction}")
```
