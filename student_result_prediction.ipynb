# Task
Predict whether a new student with 4 study hours and 50% attendance will 'Pass' or 'Fail' using the K-Nearest Neighbors (KNN) algorithm. The prediction should be based on provided training data, and the process should include loading the dataset, defining features and target, implementing KNN to predict the new student's result, visualizing the prediction with a scatter plot showing training data, the new student, and its predicted result, and finally summarizing the prediction with an explanation of the KNN steps.

## Load Dataset

### Subtask:
Load the provided student training data into a pandas DataFrame, including 'Student', 'Study Hours', 'Attendance (%)', and 'Result'.


**Reasoning**:
The subtask is to load the provided student training data into a pandas DataFrame. I will create a dictionary with the specified column names and data, then convert it into a DataFrame named `df_students`, and finally display its first few rows to verify the loading.



```python
import pandas as pd

data = {
    'Student': ['Student 1', 'Student 2', 'Student 3', 'Student 4', 'Student 5', 'Student 6', 'Student 7', 'Student 8'],
    'Study Hours': [3, 2, 5, 4, 1, 6, 3, 4],
    'Attendance (%)': [70, 40, 80, 60, 30, 90, 50, 75],
    'Result': ['Pass', 'Fail', 'Pass', 'Pass', 'Fail', 'Pass', 'Fail', 'Pass']
}

df_students = pd.DataFrame(data)

print("Student training data loaded into df_students:")
df_students.head()
```

## Define Features and Target

### Subtask:
Separate the features ('Study Hours', 'Attendance (%)') from the target variable ('Result') in the training dataset. Also, define the new student's features (Study Hours = 4, Attendance = 50%).


**Reasoning**:
I will separate the features and target variables for the training data and define the new student's features as instructed. This involves selecting specific columns from the `df_students` DataFrame for `X_train` and `y_train`, and then creating a list for `new_student_features`.



```python
X_train = df_students[['Study Hours', 'Attendance (%)']]
y_train = df_students['Result']

new_student_features = [4, 50]

print("X_train (Features):")
print(X_train.head())
print("\ny_train (Target):")
print(y_train.head())
print("\nNew student features:")
print(new_student_features)
```

## Implement KNN

### Subtask:
Implement the K-Nearest Neighbors algorithm. This will involve calculating the Euclidean distance between the new student and all training students, finding the 'k' nearest neighbors (e.g., k=3), and predicting the result based on the majority class among these neighbors.


**Reasoning**:
To implement the KNN algorithm, I need to define a value for 'k', calculate the Euclidean distances between the new student and all training students, and then combine these distances with the training results. I will start by importing the necessary `euclidean` function from `scipy.spatial.distance` and `numpy` for array operations, define `k`, and then calculate the distances, storing them in a new DataFrame along with the results.



```python
from scipy.spatial.distance import euclidean
import numpy as np

# 1. Define a value for 'k'
k = 3

# Convert new_student_features to a NumPy array for distance calculation
new_student_np = np.array(new_student_features)

# 2. Calculate the Euclidean distance between the new student and each student in X_train
distances = [
    euclidean(new_student_np, X_train.iloc[i].values)
    for i in range(len(X_train))
]

# 3. Create a DataFrame that combines the calculated distances with the corresponding 'Result' from y_train
distances_df = pd.DataFrame({
    'Student_Index': X_train.index,
    'Distance': distances,
    'Result': y_train
})

print(f"Defined k = {k}")
print("Distances and Results DataFrame (first 5 rows):")
print(distances_df.head())
```

**Reasoning**:
To identify the nearest neighbors, I will sort the `distances_df` by 'Distance' in ascending order, select the top 'k' rows, and then determine the most frequent 'Result' among these neighbors to predict the new student's outcome.



```python
sorted_distances = distances_df.sort_values(by='Distance')

# 5. Select the top 'k' rows which represent the 'k' nearest neighbors
k_nearest_neighbors = sorted_distances.head(k)

print(f"\n{k} Nearest Neighbors:")
print(k_nearest_neighbors)

# 6. Determine the majority class ('Pass' or 'Fail') among these 'k' neighbors
predicted_result = k_nearest_neighbors['Result'].mode()[0]

print(f"\nPredicted result for the new student: {predicted_result}")
```

## Visualize Prediction

### Subtask:
Generate a scatter plot to visualize the training data, clearly distinguishing between 'Pass' and 'Fail' students. Also, mark the position of the new student and indicate its predicted result with a legend.


**Reasoning**:
I will import the necessary libraries (`matplotlib.pyplot` and `seaborn`), create a combined DataFrame for plotting, then generate a scatter plot to visualize the training data and the new student's predicted result, complete with labels, title, and legend, as per the instructions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a new DataFrame by concatenating X_train and y_train for plotting
df_plot = pd.concat([X_train, y_train], axis=1)

plt.figure(figsize=(10, 7))

# Plot training data
sns.scatterplot(
    x='Study Hours',
    y='Attendance (%)',
    hue='Result',
    data=df_plot,
    s=100, # size of the points
    alpha=0.8, # transparency
    palette={'Pass': 'green', 'Fail': 'red'}
)

# Add the new student's data point
plt.scatter(
    new_student_features[0],
    new_student_features[1],
    color='blue',
    marker='X',
    s=200, # size of the marker
    label=f'New Student (Predicted: {predicted_result})'
)

plt.title('KNN Prediction for New Student')
plt.xlabel('Study Hours')
plt.ylabel('Attendance (%)')
plt.legend()
plt.grid(True)
plt.show()
```
